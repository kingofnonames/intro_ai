### Chương trình
#### 1. Giới thiệu
Chương trình mô phỏng **tìm đường trong mê cung** với hai phần chính:

- **Sinh mê cung:** sử dụng thuật toán **DFS**.  
- **Tìm đường:** cài đặt các thuật toán **DFS, BFS, DLS, IDS, UCS**.

#### 2. Thuật toán sinh mê cung

##### DFS (Depth-First Search)
- Duyệt sâu từ ô bắt đầu, chọn ngẫu nhiên một ô kề chưa thăm, phá tường và đi tiếp.  
- Khi không còn ô chưa thăm → quay lui.  
- Kết quả: mê cung có **một đường duy nhất giữa hai điểm**.

#### 3. Thuật toán tìm đường đã cài đặt

##### DFS (Depth-First Search)
- Sử dụng **ngăn xếp (Stack)** để duyệt sâu.  
- Tìm đường nhưng **không đảm bảo ngắn nhất**.

##### BFS (Breadth-First Search)
- Sử dụng **hàng đợi (Queue)** để duyệt theo tầng.  
- Đảm bảo **đường đi ngắn nhất** đến đích.

##### DLS (Depth-Limited Search)
- Biến thể của DFS có **giới hạn độ sâu**.  
- Tránh lặp vô hạn, nhưng có thể **bỏ lỡ nghiệm nếu ở sâu hơn giới hạn**.

##### IDS (Iterative Deepening Search)
- Gọi **DLS nhiều lần**, mỗi lần tăng giới hạn độ sâu.  
- Kết hợp ưu điểm của **DFS (bộ nhớ ít)** và **BFS (đảm bảo tối ưu)**.

##### UCS (Uniform Cost Search)
- Duyệt theo **chi phí nhỏ nhất** (dùng hàng đợi ưu tiên).  
- Đảm bảo **đường có chi phí tối thiểu**, phù hợp khi ô có trọng số khác nhau.

#### 4. Cấu trúc chương trình
1. Sinh mê cung bằng **DFS**.  
2. Chọn thuật toán tìm đường (DFS, BFS, DLS, IDS, UCS).  
3. Hiển thị quá trình duyệt và kết quả trực quan.

## 4. Cập nhật kết quả cuối kỳ (W15)

### Chi tiết phương pháp, dữ liệu 


#### 1. Mô hình bài toán
**Bài toán tìm đường trong mê cung (Maze Solving Problem)**

Bài toán được mô hình hóa dưới dạng **tìm đường trong mê cung**, với các đặc điểm sau:

- **Biểu diễn mê cung**: lưới 2 chiều có kích thước `rows × cols`.
- **Trạng thái**: mỗi ô trong lưới là một trạng thái.
- **Kết nối giữa các trạng thái**: các ô kề nhau được nối thông qua các cạnh theo các hướng **Trên, Dưới, Trái, Phải**.
- **Trạng thái bắt đầu**: góc trên bên trái `(0, 0)`.
- **Trạng thái đích**: góc dưới bên phải `(rows-1, cols-1)`.
- **Đảm bảo**: mỗi mê cung luôn có ít nhất **một đường đi hợp lệ** từ trạng thái bắt đầu đến trạng thái đích.

#### 2. Phương pháp sinh dữ liệu (Maze Generation)
Để đảm bảo tính đa dạng và khách quan của dữ liệu thử nghiệm, nhóm sử dụng ba thuật toán sinh mê cung khác nhau: DFS quay lui, Prim, và Kruskal. Mỗi thuật toán tạo ra mê cung với cấu trúc riêng, từ đó ảnh hưởng khác nhau đến hiệu quả của các thuật toán tìm đường.

##### 2.1 Thuật toán sinh mê cung bằng DFS
Thuật toán sinh mê cung bằng **DFS quay lui (Recursive Backtracking)** hoạt động như sau:

1. Bắt đầu từ ô `(0,0)`.
2. Liên tục mở đường sang một ô kề **chưa được thăm**, sử dụng **ngăn xếp (stack)** để ghi nhớ đường đi và quay lui khi không còn lựa chọn hợp lệ.
3. Mỗi ô ban đầu được bao quanh bởi bốn bức tường **(Trên, Dưới, Trái, Phải)**.
4. Tại mỗi bước, thuật toán chọn **ngẫu nhiên một ô kề chưa thăm** để phá tường.
5. Khi một ô không còn hàng xóm hợp lệ, thuật toán quay lui về ô trước đó.

**Đặc điểm mê cung sinh ra:**

- Đường đi dài, liên tục.  
- Ít nhánh rẽ.  
- Cấu trúc dạng **“hành lang sâu”**.  

**Nhận xét:**  
Loại mê cung này thường bất lợi cho các thuật toán tìm kiếm theo chiều sâu, vì dễ dẫn đến việc đi rất sâu trước khi gặp ngã rẽ hoặc đích, làm tăng **thời gian tìm kiếm** trong trường hợp không may.

##### 2.2 Thuật toán sinh mê cung bằng Prim
Thuật toán **Prim** hoạt động theo cơ chế **mở rộng dần từ tập frontier**. Cụ thể:

1. Khởi đầu từ một ô ban đầu và đánh dấu là đã thăm.
2. Tạo danh sách các **cạnh biên (frontiers)** nối từ ô đã thăm sang ô chưa thăm.
3. Mỗi vòng lặp, chọn **ngẫu nhiên một cạnh trong frontier** để mở rộng mê cung.
4. Lặp lại cho đến khi tất cả các ô được thăm.

**Đặc điểm mê cung sinh ra:**

- Nhiều nhánh rẽ.  
- Độ phân nhánh cao.  
- Cấu trúc cân bằng giữa chiều **dài** và chiều **rộng**.

**Nhận xét:**  
Mê cung kiểu này đặc biệt ảnh hưởng đến **bộ nhớ của BFS**, do BFS cần lưu trữ nhiều trạng thái ở cùng một mức độ sâu trong quá trình tìm kiếm.

##### 2.3 Thuật toán sinh mê cung bằng Kruskal

Thuật toán **Kruskal** dựa trên **cấu trúc tập hợp rời (Union-Find)**. Cụ thể:

1. Mỗi ô trong mê cung được xem như một **tập hợp độc lập** ban đầu.
2. Mỗi ô được ánh xạ thành một **chỉ số duy nhất**.
3. Các cạnh tiềm năng giữa các ô kề nhau được **liệt kê và xáo trộn ngẫu nhiên**.
4. Một cạnh chỉ được **phá tường** nếu hai ô thuộc **hai tập khác nhau** (theo Union-Find), đảm bảo **không tạo chu trình**.

**Đặc điểm mê cung sinh ra:**

- Phân bố đường đi khá **đồng đều**.  
- Ít thiên lệch về **chiều sâu hoặc chiều rộng**.  
- **Không tạo chu trình**, đảm bảo chỉ có **một đường đi duy nhất** giữa hai ô.

**Nhận xét:**  
Mê cung kiểu này mang tính **trung tính**, phù hợp cho việc **so sánh tổng quát hiệu năng** của các thuật toán tìm đường khác nhau.

#### 3. Các thuật toán tìm đường được đánh giá
Nhóm tiến hành đánh giá các thuật toán tìm kiếm không sử dụng thông tin heuristic và sử dụng thông tin heuristic.

##### 3.1. Thuật toán không sử dụng heuristic

**Breadth-First Search (BFS)**
- Duyệt theo từng tầng từ ô bắt đầu, sử dụng hàng đợi (queue) để lưu các ô cần khám phá.
- Đảm bảo tìm ra đường đi ngắn nhất đến đích khi chi phí đồng nhất.
- Ghi lại thứ tự duyệt, cha của từng ô, độ sâu và kích thước frontier.

**Depth-First Search (DFS)**
- Duyệt sâu từ ô bắt đầu, sử dụng ngăn xếp (stack) để ghi nhớ đường đi.
- Không đảm bảo đường đi ngắn nhất.
- Ghi lại thứ tự duyệt, cha của từng ô, độ sâu và kích thước frontier.

**Depth-Limited Search (DLS)**
- Biến thể của DFS với giới hạn độ sâu.
- Tránh lặp vô hạn nhưng có thể bỏ lỡ nghiệm nếu nằm ngoài giới hạn.

**Iterative Deepening Search (IDS)**
- Kết hợp DFS và BFS bằng cách gọi lặp lại DLS với giới hạn độ sâu tăng dần.
- Giữ ưu điểm bộ nhớ ít của DFS và đảm bảo tìm được đường đi tối ưu như BFS** trong trường hợp chi phí đồng nhất.

**Uniform Cost Search (UCS / UCB)**
- Duyệt theo chi phí tích lũy thấp nhất sử dụng hàng đợi ưu tiên.
- Trong bài toán không trọng số (mọi ô bằng nhau), UCS tương đương BFS.

##### 3.2. Thuật toán sử dụng heuristic

**A \***
- Duyệt các ô dựa trên hàm `f(n) = g(n) + h(n)`, trong đó:
  - `g(n)` là chi phí từ ô bắt đầu đến ô hiện tại.
  - `h(n)` là heuristic, ở đây sử dụng **Manhattan distance** đến đích:
  
  \[
  h((r, c), goal) = |r - goal_r| + |c - goal_c|
  \]
- Ưu điểm: đảm bảo tìm được đường đi ngắn nhất và tiết kiệm thời gian khám phá nhờ thông tin heuristic.

**Greedy Best-First Search (GBFS)**
- Duyệt các ô dựa trên heuristic `h(n)` mà không tính chi phí từ bắt đầu (`g(n)`).
- Sử dụng **Manhattan distance** để đánh giá mức “gần đích” của mỗi ô.
- Ưu điểm: nhanh trong nhiều trường hợp.  
- Nhược điểm: không đảm bảo đường đi tối ưu.


##### Ghi chú chung về cài đặt

- **Các thuật toán** đều lưu lại các thông tin:
  - **Thứ tự ô được duyệt** (`order`)
  - **Cha của ô** (`parent`)
  - **Độ sâu** (`depth`)
  - **Kích thước frontier**

- **Mục đích:** thông tin này giúp đánh giá **hiệu quả**, **bộ nhớ sử dụng** và đặc điểm **quá trình duyệt** của từng thuật toán.

3. **Thu thập dữ liệu** trong quá trình tìm kiếm.
4. **Lặp lại mỗi cấu hình 3 lần** để giảm ảnh hưởng của yếu tố ngẫu nhiên.
5. **Lưu toàn bộ kết quả** để phục vụ trực quan hóa và phân tích.

```bash
cd intro_ai
pip install -r requirements.txt
python main.py

### Phân tích, đánh giá kết quả

#### 1. Kết quả đạt được với thuật toán sinh mê cung dựa trên DFS

| Kích thước maze | Thuật toán | Tổng node duyệt (Mean ± Std) | Độ sâu tối đa (Mean ± Std) | Số node lưu trữ lớn nhất (Mean ± Std) |
| --------------- | ---------- | ---------------------------- | -------------------------- | ------------------------------------- |
| 5x5             | bfs        | 18 ± 4.546                   | 13.333 ± 1.886             | 2 ± 0.816                             |
| 5x5             | dfs        | 17.667 ± 4.11                | 13.333 ± 1.886             | 1.667 ± 0.471                         |
| 5x5             | ids        | 123.333 ± 37.677             | 13.333 ± 1.886             | 1.667 ± 0.471                         |
| 5x5             | dls        | 12 ± 1.414                   | 10 ± 0                     | 1.333 ± 0.471                         |
| 5x5             | ucb        | 18 ± 4.546                   | 13.333 ± 1.886             | 2 ± 0.816                             |
| 5x5             | astar      | 15.333 ± 2.625               | 13.333 ± 1.886             | 2.667 ± 1.247                         |
| 5x5             | gbfs       | 14.667 ± 2.357               | 13.333 ± 1.886             | 2.667 ± 1.247                         |
| 10x10           | bfs        | 67 ± 14.765                  | 48.667 ± 8.055             | 3 ± 0.816                             |
| 10x10           | dfs        | 56 ± 13.441                  | 48.667 ± 8.055             | 3.667 ± 0.943                         |
| 10x10           | ids        | 1593.333 ± 673.065           | 48.667 ± 8.055             | 3.667 ± 0.943                         |
| 10x10           | dls        | 11.667 ± 0.943               | 10 ± 0                     | 1.333 ± 0.471                         |
| 10x10           | ucb        | 67 ± 14.765                  | 48.667 ± 8.055             | 3 ± 0.816                             |
| 10x10           | astar      | 60.333 ± 15.37               | 48.667 ± 8.055             | 3 ± 0.816                             |
| 10x10           | gbfs       | 55.667 ± 15.107              | 48.667 ± 8.055             | 4.667 ± 1.247                         |
| 20x20           | bfs        | 274 ± 106.104                | 163 ± 43.043               | 5 ± 1.633                             |
| 20x20           | dfs        | 189.667 ± 62.404             | 163 ± 43.043               | 9 ± 2.449                             |
| 20x20           | ids        | 20526 ± 9879.439             | 162.667 ± 42.968           | 9 ± 2.449                             |
| 20x20           | dls        | 11 ± 0                       | 10 ± 0                     | 1 ± 0                                 |
| 20x20           | ucb        | 274 ± 106.104                | 163 ± 43.043               | 5 ± 1.633                             |
| 20x20           | astar      | 256.333 ± 97.044             | 162.667 ± 42.968           | 6 ± 2.944                             |
| 20x20           | gbfs       | 199 ± 63.124                 | 162.667 ± 42.968           | 7.667 ± 1.886                         |
| 30x30           | bfs        | 435 ± 200.833                | 270.333 ± 84.941           | 7.667 ± 4.497                         |
| 30x30           | dfs        | 333.667 ± 116.691            | 275.667 ± 92.168           | 16 ± 7.483                            |
| 30x30           | ids        | 55543.667 ± 35598.387        | 270 ± 84.491               | 14.667 ± 5.735                        |
| 30x30           | dls        | 11 ± 0                       | 10 ± 0                     | 1 ± 0                                 |
| 30x30           | ucb        | 435 ± 200.833                | 270.333 ± 84.941           | 7.667 ± 4.497                         |
| 30x30           | astar      | 384.667 ± 147.083            | 270 ± 84.491               | 5.333 ± 1.247                         |
| 30x30           | gbfs       | 314.333 ± 101.065            | 270 ± 84.491               | 14 ± 5.354                            |
| 40x40           | bfs        | 889.333 ± 449.188            | 550 ± 264.939              | 9 ± 3.742                             |
| 40x40           | dfs        | 1044.667 ± 254.326           | 684 ± 197.916              | 30 ± 9.899                            |
| 40x40           | ids        | 280548 ± 277796.835          | 549.333 ± 265.402          | 23 ± 14.967                           |
| 40x40           | dls        | 12 ± 1.414                   | 10 ± 0                     | 1.333 ± 0.471                         |
| 40x40           | ucb        | 889.333 ± 449.188            | 550 ± 264.939              | 9 ± 3.742                             |
| 40x40           | astar      | 827.333 ± 480.036            | 549.333 ± 265.402          | 6.333 ± 1.7                           |
| 40x40           | gbfs       | 674.667 ± 379.574            | 549.333 ± 265.402          | 19 ± 6.532                            |
| 50x50           | bfs        | 1806.333 ± 407.019           | 949 ± 140.057              | 8.333 ± 1.247                         |
| 50x50           | dfs        | 1883 ± 342.959               | 1148.333 ± 72.039          | 47.333 ± 5.185                        |
| 50x50           | ids        | 751699 ± 226539.126          | 948 ± 140.057              | 36.333 ± 9.393                        |
| 50x50           | dls        | 11.333 ± 0.471               | 10 ± 0                     | 1.333 ± 0.471                         |
| 50x50           | ucb        | 1806.333 ± 407.019           | 949 ± 140.057              | 8.333 ± 1.247                         |
| 50x50           | astar      | 1695 ± 356.458               | 948 ± 140.057              | 8.333 ± 0.943                         |
| 50x50           | gbfs       | 1351 ± 346.394               | 948 ± 140.057              | 23.667 ± 0.943                        |
| 70x70           | bfs        | 3278.667 ± 463.7             | 1512.333 ± 313.655         | 13 ± 2.16                             |
| 70x70           | dfs        | 3487 ± 403.95                | 1714.333 ± 134.068         | 73 ± 12.356                           |
| 70x70           | ids        | 2135232.333 ± 799616.792     | 1511.333 ± 313.655         | 63.333 ± 19.939                       |
| 70x70           | dls        | 11 ± 0                       | 10 ± 0                     | 1 ± 0                                 |
| 70x70           | ucb        | 3278.667 ± 463.7             | 1512.333 ± 313.655         | 13 ± 2.16                             |
| 70x70           | astar      | 3037.667 ± 493.632           | 1511.333 ± 313.655         | 10.667 ± 1.7                          |
| 70x70           | gbfs       | 2108.333 ± 446.139           | 1511.333 ± 313.655         | 25 ± 3.559                            |
| 90x90           | bfs        | 3681.667 ± 882.746           | 1845.667 ± 464.945         | 11 ± 1.414                            |
| 90x90           | dfs        | 4705.333 ± 1805.507          | 2659.667 ± 437.741         | 100 ± 16.31                           |
| 90x90           | ids        | 3446081.667 ± 1550077.508    | 1844.667 ± 464.945         | 65.667 ± 8.38                         |
| 90x90           | dls        | 11.333 ± 0.471               | 10 ± 0                     | 1.333 ± 0.471                         |
| 90x90           | ucb        | 3681.667 ± 882.746           | 1845.667 ± 464.945         | 11 ± 1.414                            |
| 90x90           | astar      | 3564.667 ± 825.161           | 1844.667 ± 464.945         | 13 ± 0.816                            |
| 90x90           | gbfs       | 2391 ± 775.86                | 1844.667 ± 464.945         | 33 ± 4.243                            |
| 100x100         | bfs        | 6039 ± 1013.506              | 2762.667 ± 420.735         | 10.667 ± 1.247                        |
| 100x100         | dfs        | 4195 ± 753.794               | 2878 ± 386.612             | 104.667 ± 10.209                      |
| 100x100         | ids        | 7739022.667 ± 2280283.257    | 2762 ± 421.201             | 99.333 ± 12.499                       |
| 100x100         | dls        | 11.333 ± 0.471               | 10 ± 0                     | 1.333 ± 0.471                         |
| 100x100         | ucb        | 6039 ± 1013.506              | 2762.667 ± 420.735         | 10.667 ± 1.247                        |
| 100x100         | astar      | 5898.667 ± 1110.76           | 2762 ± 421.201             | 12 ± 0.816                            |
| 100x100         | gbfs       | 4175 ± 861.136               | 2762 ± 421.201             | 56.667 ± 4.989                        |
| 120x120         | bfs        | 10400 ± 1920.782             | 3930.333 ± 719.365         | 15.333 ± 0.471                        |
| 120x120         | dfs        | 9622.333 ± 1401.805          | 4512.333 ± 193.882         | 163.667 ± 2.357                       |
| 120x120         | ids        | 15882629 ± 5837228.165       | 3929.333 ± 719.365         | 138.333 ± 21.746                      |
| 120x120         | dls        | 11.667 ± 0.943               | 10 ± 0                     | 1.333 ± 0.471                         |
| 120x120         | ucb        | 10400 ± 1920.782             | 3930.333 ± 719.365         | 15.333 ± 0.471                        |
| 120x120         | astar      | 9874.333 ± 1976.504          | 3929.333 ± 719.365         | 16.333 ± 1.7                          |
| 120x120         | gbfs       | 6861 ± 2083.953              | 3929.333 ± 719.365         | 50.333 ± 1.247                        |
| 150x150         | bfs        | 15316 ± 5546.957             | 5748.667 ± 1922.422        | 15.667 ± 1.247                        |
| 150x150         | dfs        | 10624.333 ± 2396.044         | 6264 ± 1512.927            | 220.667 ± 54.365                      |
| 150x150         | ids        | 41444993.33 ± 23107421.15    | 5748 ± 1922.782            | 199 ± 64.545                          |
| 150x150         | dls        | 11 ± 0                       | 10 ± 0                     | 1 ± 0                                 |
| 150x150         | ucb        | 15316 ± 5546.957             | 5748.667 ± 1922.422        | 15.667 ± 1.247                        |
| 150x150         | astar      | 14774.333 ± 5336.711         | 5748 ± 1922.782            | 16 ± 2.16                             |
| 150x150         | gbfs       | 9403 ± 4115.93               | 5748 ± 1922.782            | 60.667 ± 6.342                        |


#### 2. Kết quả đạt được với thuật toán sinh mê cung dựa trên Prim

| Kích thước maze | Thuật toán | Tổng node duyệt (Mean ± Std) | Độ sâu tối đa (Mean ± Std) | Số node lưu trữ lớn nhất (Mean ± Std) |
| --------------- | ---------- | ---------------------------- | -------------------------- | ------------------------------------- |
| 5x5             | bfs        | 25 ± 0                       | 8.667 ± 0.943              | 5.333 ± 0.471                         |
| 5x5             | dfs        | 17.333 ± 2.055               | 8.667 ± 0.943              | 4.333 ± 0.471                         |
| 5x5             | ids        | 116.333 ± 20.434             | 8.667 ± 0.943              | 4.333 ± 0.471                         |
| 5x5             | dls        | 17.333 ± 2.055               | 8.667 ± 0.943              | 4.333 ± 0.471                         |
| 5x5             | ucb        | 25 ± 0                       | 8.667 ± 0.943              | 5.333 ± 0.471                         |
| 5x5             | astar      | 20.667 ± 4.784               | 8.667 ± 0.943              | 5.333 ± 0.471                         |
| 5x5             | gbfs       | 16 ± 5.099                   | 8.667 ± 0.943              | 5 ± 1.414                             |
| 10x10           | bfs        | 100 ± 0                      | 20.667 ± 0.943             | 10.667 ± 0.471                        |
| 10x10           | dfs        | 66.667 ± 14.614              | 20.667 ± 0.943             | 11.667 ± 0.943                        |
| 10x10           | ids        | 1042.333 ± 57.25             | 20.667 ± 0.943             | 11.667 ± 0.943                        |
| 10x10           | dls        | 44 ± 3.559                   | 10 ± 0                     | 9.667 ± 2.494                         |
| 10x10           | ucb        | 100 ± 0                      | 20.667 ± 0.943             | 10.667 ± 0.471                        |
| 10x10           | astar      | 82 ± 5.715                   | 20.667 ± 0.943             | 13.667 ± 1.886                        |
| 10x10           | gbfs       | 50 ± 12.193                  | 20.667 ± 0.943             | 10.333 ± 1.886                        |
| 20x20           | bfs        | 398.667 ± 1.247              | 39 ± 0.816                 | 20.667 ± 1.7                          |
| 20x20           | dfs        | 174.667 ± 59.298             | 39 ± 0.816                 | 17 ± 2.944                            |
| 20x20           | ids        | 6626 ± 394.439               | 38.667 ± 0.943             | 17.667 ± 2.055                        |
| 20x20           | dls        | 46 ± 7.071                   | 10 ± 0                     | 6 ± 0                                 |
| 20x20           | ucb        | 398.667 ± 1.247              | 39 ± 0.816                 | 20.667 ± 1.7                          |
| 20x20           | astar      | 147.333 ± 40.434             | 38.667 ± 0.943             | 31.667 ± 7.409                        |
| 20x20           | gbfs       | 206.333 ± 48.917             | 38.667 ± 0.943             | 22.333 ± 4.784                        |
| 30x30           | bfs        | 897.333 ± 1.7                | 61.667 ± 1.247             | 33 ± 2.449                            |
| 30x30           | dfs        | 743.333 ± 48.61              | 62 ± 1.633                 | 28 ± 2.16                             |
| 30x30           | ids        | 25611.333 ± 353.336          | 61.333 ± 0.943             | 28 ± 2.16                             |
| 30x30           | dls        | 46.667 ± 1.886               | 10 ± 0                     | 7 ± 1.414                             |
| 30x30           | ucb        | 897.333 ± 1.7                | 61.667 ± 1.247             | 33 ± 2.449                            |
| 30x30           | astar      | 488.333 ± 49.304             | 61.333 ± 0.943             | 64.667 ± 6.944                        |
| 30x30           | gbfs       | 167 ± 43.459                 | 61.333 ± 0.943             | 37.333 ± 3.771                        |
| 40x40           | bfs        | 1599 ± 1.414                 | 82 ± 3.266                 | 42.667 ± 5.735                        |
| 40x40           | dfs        | 991.333 ± 298.602            | 82 ± 3.266                 | 34.667 ± 1.247                        |
| 40x40           | ids        | 59743 ± 2218.081             | 82 ± 3.266                 | 34.667 ± 1.247                        |
| 40x40           | dls        | 48.333 ± 4.028               | 10 ± 0                     | 7.333 ± 1.247                         |
| 40x40           | ucb        | 1599 ± 1.414                 | 82 ± 3.266                 | 42.667 ± 5.735                        |
| 40x40           | astar      | 720.333 ± 168.359            | 82 ± 3.266                 | 80.667 ± 13.474                       |
| 40x40           | gbfs       | 623 ± 190.354                | 82 ± 3.266                 | 49.667 ± 4.922                        |
| 50x50           | bfs        | 2495.333 ± 2.867             | 101 ± 2.16                 | 59.667 ± 1.7                          |
| 50x50           | dfs        | 1613.333 ± 447.773           | 101.333 ± 1.886            | 44.333 ± 1.7                          |
| 50x50           | ids        | 112105 ± 3603.74             | 100.667 ± 2.494            | 44.333 ± 1.7                          |
| 50x50           | dls        | 45.667 ± 12.392              | 10 ± 0                     | 5.333 ± 1.247                         |
| 50x50           | ucb        | 2495.333 ± 2.867             | 101 ± 2.16                 | 59.667 ± 1.7                          |
| 50x50           | astar      | 791.333 ± 297.564            | 100.667 ± 2.494            | 106.667 ± 18.264                      |
| 50x50           | gbfs       | 749 ± 312.347                | 101 ± 2.16                 | 71.667 ± 7.76                         |
| 70x70           | bfs        | 4899.333 ± 0.471             | 145.667 ± 3.859            | 79 ± 5.888                            |
| 70x70           | dfs        | 1951.667 ± 730.378           | 145.667 ± 3.859            | 60.333 ± 3.859                        |
| 70x70           | ids        | 318449.667 ± 12131.085       | 145.333 ± 3.399            | 61 ± 2.944                            |
| 70x70           | dls        | 48 ± 3.559                   | 10 ± 0                     | 6 ± 0                                 |
| 70x70           | ucb        | 4899.333 ± 0.471             | 145.667 ± 3.859            | 79 ± 5.888                            |
| 70x70           | astar      | 1911.667 ± 656.73            | 145.333 ± 3.399            | 148 ± 11.576                          |
| 70x70           | gbfs       | 2623.333 ± 697.871           | 145.333 ± 3.399            | 87 ± 6.683                            |
| 90x90           | bfs        | 8096.333 ± 2.625             | 186.667 ± 0.471            | 105.333 ± 5.249                       |
| 90x90           | dfs        | 3274.667 ± 1443.568          | 187.667 ± 1.7              | 84.333 ± 2.494                        |
| 90x90           | ids        | 675198.333 ± 10217.166       | 186 ± 0                    | 84.333 ± 2.494                        |
| 90x90           | dls        | 44.667 ± 9.568               | 10 ± 0                     | 7.667 ± 2.494                         |
| 90x90           | ucb        | 8096.333 ± 2.625             | 186.667 ± 0.471            | 105.333 ± 5.249                       |
| 90x90           | astar      | 2921.667 ± 605.703           | 186 ± 0                    | 234.333 ± 27.354                      |
| 90x90           | gbfs       | 3998 ± 900.813               | 186 ± 0                    | 113 ± 1.414                           |
| 100x100         | bfs        | 9998.333 ± 1.247             | 214.667 ± 3.771            | 105 ± 4.899                           |
| 100x100         | dfs        | 4766 ± 1020.422              | 214.667 ± 3.771            | 96.667 ± 7.364                        |
| 100x100         | ids        | 1013911 ± 13719.874          | 214.667 ± 3.771            | 96.667 ± 7.364                        |
| 100x100         | dls        | 41.667 ± 10.625              | 10 ± 0                     | 7 ± 0.816                             |
| 100x100         | ucb        | 9998.333 ± 1.247             | 214.667 ± 3.771            | 105 ± 4.899                           |
| 100x100         | astar      | 6784.333 ± 645.413           | 214.667 ± 3.771            | 270 ± 33.347                          |
| 100x100         | gbfs       | 4860.333 ± 967.685           | 214.667 ± 3.771            | 132.667 ± 2.055                       |
| 120x120         | bfs        | 14399 ± 0.816                | 261.333 ± 9.286            | 141.667 ± 10.339                      |
| 120x120         | dfs        | 7893.667 ± 4501.707          | 261.333 ± 9.286            | 104.667 ± 9.104                       |
| 120x120         | ids        | 1779503.667 ± 106677.539     | 261.333 ± 9.286            | 111.333 ± 1.7                         |
| 120x120         | dls        | 47.667 ± 9.104               | 10 ± 0                     | 9.333 ± 2.055                         |
| 120x120         | ucb        | 14399 ± 0.816                | 261.333 ± 9.286            | 141.667 ± 10.339                      |
| 120x120         | astar      | 9957.333 ± 2428.858          | 261.333 ± 9.286            | 293.333 ± 31.33                       |
| 120x120         | gbfs       | 7052.667 ± 4069.259          | 261.333 ± 9.286            | 145 ± 5.354                           |
| 150x150         | bfs        | 22496.667 ± 2.357            | 318.333 ± 9.031            | 166.667 ± 9.286                       |
| 150x150         | dfs        | 8851.667 ± 1136.358          | 319.333 ± 8.994            | 134 ± 8.042                           |
| 150x150         | ids        | 3323285.333 ± 149904.189     | 318 ± 9.092                | 135 ± 7.118                           |
| 150x150         | dls        | 46 ± 5.888                   | 10 ± 0                     | 6.667 ± 1.247                         |
| 150x150         | ucb        | 22496.667 ± 2.357            | 318.333 ± 9.031            | 166.667 ± 9.286                       |
| 150x150         | astar      | 12880.333 ± 4028.207         | 318 ± 9.092                | 446.667 ± 29.409                      |
| 150x150         | gbfs       | 9938 ± 802.368               | 318.333 ± 8.731            | 177.667 ± 6.128                       |


#### 3. Kết quả đạt được với thuật toán sinh mê cung dựa trên Kruskal

| Kích thước maze | Thuật toán | Tổng node duyệt (Mean ± Std) | Độ sâu tối đa (Mean ± Std) | Số node lưu trữ lớn nhất (Mean ± Std) |
| --------------- | ---------- | ---------------------------- | -------------------------- | ------------------------------------- |
| 5x5             | bfs        | 20 ± 3.266                   | 8.333 ± 0.471              | 4.667 ± 0.471                         |
| 5x5             | dfs        | 16 ± 2.449                   | 8.667 ± 0.943              | 3.667 ± 0.471                         |
| 5x5             | ids        | 86.333 ± 14.384              | 8 ± 0                      | 3.333 ± 0.471                         |
| 5x5             | dls        | 16 ± 2.449                   | 8.667 ± 0.943              | 3.667 ± 0.471                         |
| 5x5             | ucb        | 20 ± 3.266                   | 8.333 ± 0.471              | 4.667 ± 0.471                         |
| 5x5             | astar      | 15.667 ± 2.625               | 8 ± 0                      | 4.333 ± 0.471                         |
| 5x5             | gbfs       | 13.333 ± 2.494               | 8 ± 0                      | 4.667 ± 0.943                         |
| 10x10           | bfs        | 86.333 ± 11.086              | 27 ± 2.828                 | 8.333 ± 1.247                         |
| 10x10           | dfs        | 82 ± 12.57                   | 31.333 ± 3.682             | 9.333 ± 0.471                         |
| 10x10           | ids        | 1045.667 ± 253.214           | 26 ± 2.828                 | 8 ± 1.414                             |
| 10x10           | dls        | 24.333 ± 5.437               | 10 ± 0                     | 4.333 ± 0.943                         |
| 10x10           | ucb        | 86.333 ± 11.086              | 27 ± 2.828                 | 8.333 ± 1.247                         |
| 10x10           | astar      | 68.333 ± 10.873              | 26 ± 2.828                 | 9.333 ± 1.247                         |
| 10x10           | gbfs       | 47 ± 0.816                   | 26 ± 2.828                 | 13 ± 2.944                            |
| 20x20           | bfs        | 350.333 ± 17.594             | 53.333 ± 5.793             | 15.333 ± 1.7                          |
| 20x20           | dfs        | 232 ± 38.944                 | 58.667 ± 6.799             | 22.333 ± 0.943                        |
| 20x20           | ids        | 7315.333 ± 841.715           | 52.667 ± 5.735             | 20.667 ± 1.247                        |
| 20x20           | dls        | 24.667 ± 6.6                 | 10 ± 0                     | 4.667 ± 0.943                         |
| 20x20           | ucb        | 350.333 ± 17.594             | 53.333 ± 5.793             | 15.333 ± 1.7                          |
| 20x20           | astar      | 199.333 ± 19.956             | 52.667 ± 5.735             | 26 ± 4.32                             |
| 20x20           | gbfs       | 127.333 ± 34.422             | 52.667 ± 5.735             | 32.667 ± 4.028                        |
| 30x30           | bfs        | 751.333 ± 89.108             | 91 ± 13.952                | 22.333 ± 4.028                        |
| 30x30           | dfs        | 479 ± 162.236                | 101.667 ± 16.78            | 29.667 ± 3.859                        |
| 30x30           | ids        | 29568.667 ± 9590.838         | 90 ± 13.952                | 28.333 ± 3.682                        |
| 30x30           | dls        | 21.667 ± 4.922               | 10 ± 0                     | 5 ± 0.816                             |
| 30x30           | ucb        | 751.333 ± 89.108             | 91 ± 13.952                | 22.333 ± 4.028                        |
| 30x30           | astar      | 516.667 ± 182.46             | 90 ± 13.952                | 27 ± 1.414                            |
| 30x30           | gbfs       | 166.667 ± 53.481             | 90 ± 13.952                | 35.667 ± 1.7                          |
| 40x40           | bfs        | 1425.667 ± 131.566           | 130 ± 16.062               | 37 ± 2.449                            |
| 40x40           | dfs        | 755.667 ± 402.126            | 145.333 ± 6.799            | 43.333 ± 5.185                        |
| 40x40           | ids        | 69464 ± 19272.954            | 129.333 ± 16.438           | 42 ± 4.082                            |
| 40x40           | dls        | 26 ± 5.715                   | 10 ± 0                     | 4.667 ± 0.943                         |
| 40x40           | ucb        | 1425.667 ± 131.566           | 130 ± 16.062               | 37 ± 2.449                            |
| 40x40           | astar      | 996.667 ± 234.153            | 129.333 ± 16.438           | 43.333 ± 6.65                         |
| 40x40           | gbfs       | 873.667 ± 151.002            | 134.333 ± 15.521           | 54.333 ± 1.886                        |
| 50x50           | bfs        | 1927.333 ± 308.815           | 171.333 ± 12.97            | 28.333 ± 4.714                        |
| 50x50           | dfs        | 1714 ± 285.455               | 209.667 ± 28.987           | 66.667 ± 11.585                       |
| 50x50           | ids        | 147072.667 ± 39014.407       | 170.667 ± 13.199           | 56.333 ± 2.055                        |
| 50x50           | dls        | 20 ± 2.16                    | 10 ± 0                     | 4 ± 0                                 |
| 50x50           | ucb        | 1927.333 ± 308.815           | 171.333 ± 12.97            | 28.333 ± 4.714                        |
| 50x50           | astar      | 1600.333 ± 361.221           | 170.667 ± 13.199           | 41.667 ± 2.055                        |
| 50x50           | gbfs       | 729.333 ± 277.71             | 172.333 ± 14.522           | 79 ± 5.099                            |
| 70x70           | bfs        | 4221.667 ± 257.118           | 309.333 ± 14.614           | 37.667 ± 1.7                          |
| 70x70           | dfs        | 2791.667 ± 834.259           | 354.333 ± 48.992           | 116.333 ± 15.085                      |
| 70x70           | ids        | 554479.667 ± 66083.726       | 308.667 ± 14.267           | 106.333 ± 8.014                       |
| 70x70           | dls        | 27 ± 6.683                   | 10 ± 0                     | 5.333 ± 0.471                         |
| 70x70           | ucb        | 4221.667 ± 257.118           | 309.333 ± 14.614           | 37.667 ± 1.7                          |
| 70x70           | astar      | 3293 ± 457.447               | 308.667 ± 14.267           | 48 ± 3.742                            |
| 70x70           | gbfs       | 1560.333 ± 98.611            | 308.667 ± 14.267           | 117 ± 4.32                            |
| 90x90           | bfs        | 5881.333 ± 721.047           | 368.667 ± 55.859           | 45 ± 9.416                            |
| 90x90           | dfs        | 4469 ± 1417.225              | 446.333 ± 71.041           | 154.333 ± 25.171                      |
| 90x90           | ids        | 936246.333 ± 246153.486      | 368 ± 55.809               | 133.667 ± 18.062                      |
| 90x90           | dls        | 29.667 ± 2.055               | 10 ± 0                     | 5 ± 0                                 |
| 90x90           | ucb        | 5881.333 ± 721.047           | 368.667 ± 55.859           | 45 ± 9.416                            |
| 90x90           | astar      | 4958.333 ± 815.499           | 368 ± 55.809               | 61.667 ± 3.3                          |
| 90x90           | gbfs       | 1774 ± 714.741               | 368 ± 55.809               | 154.333 ± 5.312                       |
| 100x100         | bfs        | 9674.333 ± 128.93            | 447 ± 27.905               | 71.333 ± 13.021                       |
| 100x100         | dfs        | 4243 ± 1613.409              | 453 ± 28.331               | 141 ± 13.491                          |
| 100x100         | ids        | 1746404.333 ± 71893.306      | 446 ± 27.905               | 143 ± 16.391                          |
| 100x100         | dls        | 20.667 ± 4.497               | 10 ± 0                     | 3 ± 1.414                             |
| 100x100         | ucb        | 9674.333 ± 128.93            | 447 ± 27.905               | 71.333 ± 13.021                       |
| 100x100         | astar      | 8281 ± 684.79                | 446 ± 27.905               | 85 ± 15.578                           |
| 100x100         | gbfs       | 2889.667 ± 548.089           | 446 ± 27.905               | 163.333 ± 6.182                       |
| 120x120         | bfs        | 12441.333 ± 1632.123         | 534.667 ± 71.444           | 63 ± 5.354                            |
| 120x120         | dfs        | 4869.333 ± 763.362           | 536.333 ± 70.22            | 181.333 ± 23.099                      |
| 120x120         | ids        | 2840449.667 ± 814978.568     | 534 ± 71.852               | 184 ± 25.469                          |
| 120x120         | dls        | 24.333 ± 4.11                | 10 ± 0                     | 5.333 ± 1.7                           |
| 120x120         | ucb        | 12441.333 ± 1632.123         | 534.667 ± 71.444           | 63 ± 5.354                            |
| 120x120         | astar      | 10466 ± 1826.178             | 534 ± 71.852               | 80.333 ± 13.912                       |
| 120x120         | gbfs       | 4919.333 ± 2614.144          | 534 ± 71.852               | 202 ± 9.416                           |
| 150x150         | bfs        | 19662 ± 3434.242             | 685.667 ± 44.312           | 76 ± 21.276                           |
| 150x150         | dfs        | 9706.333 ± 3918.089          | 703.333 ± 24.281           | 236.333 ± 10.625                      |
| 150x150         | ids        | 5496718 ± 813419.172         | 684.667 ± 44.312           | 233.333 ± 16.819                      |
| 150x150         | dls        | 20 ± 3.742                   | 10 ± 0                     | 6.333 ± 1.7                           |
| 150x150         | ucb        | 19662 ± 3434.242             | 685.667 ± 44.312           | 76 ± 21.276                           |
| 150x150         | astar      | 17109 ± 2385.257             | 684.667 ± 44.312           | 111 ± 26.87                           |
| 150x150         | gbfs       | 6839.333 ± 2032.103          | 684.667 ± 44.312           | 268.667 ± 10.656                      |


#### 4. Trực quan hóa kết quả
Nhóm thực hiện trực quan hóa kết quả của các thuật toán tìm đường trên các mê cung có kích thước: 5×5, 30×30, 70×70, 100×100, 150×150.
Các tiêu chí đánh giá gồm:
- Tổng số nút phải duyệt: số ô mà thuật toán đã khám phá trong quá trình tìm đường.

- Độ sâu lớn nhất đạt được: mức độ sâu tối đa của thuật toán trong quá trình duyệt.

- Kích thước frontier lớn nhất: số lượng trạng thái tối đa được lưu trong frontier tại bất kỳ thời điểm nào.

**Mê cung kích thước 5x5**




**Mê cung kích thước 30x30**




**Mê cung kích thước 70x70**




**Mê cung kích thước 100x100**




**Mê cung kích thước 150x150**




#### 5. Đánh giá kết quả

##### Nhận xét tổng quan về thuật toán tìm đường

**BFS và UCS:**
- Do tất cả các cạnh trong mê cung đều có chi phí bằng 1, nên UCS sẽ hoạt động tương tự BFS.
- BFS/UCS luôn đảm bảo tìm được đường đi ngắn nhất (shortest path) theo số bước, nhưng total_nodes tăng mạnh khi kích thước mê cung tăng.

**DFS:**
- DFS thường thăm ít nút hơn BFS khi kích thước mê cung tăng.
- Ở mê cung do Prim sinh ra, max_frontier của DFS thấp hơn BFS tuy nhiên với mê cung do Kruskal hoặc DFS sinh ra thì ngược lại. Nguyên nhân là do Prim sinh ra mê cung có nhiều nhánh, dẫn đến BFS phải duy trì một frontier lớn hơn để thăm hết các nhánh cùng một lúc. DFS, với tính chất thăm sâu, chỉ cần giữ frontier nhỏ hơn nhiều vì chỉ theo một nhánh đến cùng trước khi quay lui.
- DFS không đảm bảo tìm đường ngắn nhất, nhưng trong nhiều trường hợp thăm ít nút hơn so với BFS.

**IDS (Iterative Deepening Search)**
- IDS hoạt động cực kỳ tốn kém về total_nodes, đặc biệt với các mê cung lớn và được sinh bằng DFS (ví dụ 20x20 trở lên).
- Mặc dù IDS kết hợp ưu điểm của DFS về giới hạn độ sâu và tính chất optimal của BFS, chi phí lặp lại nhiều lần các mức độ sâu làm tăng tổng số nút thăm rất nhiều.

**DLS (Depth-Limited Search):**
- DLS chỉ thăm ít nút và frontier rất nhỏ, vì giới hạn độ sâu cố định (thường là 10).
- Tuy nhiên, DLS không đảm bảo tìm đường đi, trừ khi đường đi nằm trong giới hạn độ sâu.

**A \*:**
- A* hoạt động hiệu quả hơn BFS về tổng số nút thăm (total_nodes thấp hơn), nhờ sử dụng heuristic để hướng đến mục tiêu.
- Tuy nhiên, với các mê cung phức tạp (Prim và Kruskal, đặc biệt kích thước lớn), max_frontier có thể tăng cao do heuristic không hoàn hảo và số lượng nút trong frontier lớn.

**GBFS (Greedy Best-First Search):**
- GBFS thăm ít nút hơn BFS/DFS nhưng không đảm bảo đường đi ngắn nhất.
- max_frontier thường lớn hơn DFS nhưng nhỏ hơn BFS trong nhiều trường hợp, vì GBFS tập trung chọn nút gần mục tiêu hơn là thăm tuần tự.

##### Nhận xét theo thuật toán sinh mê cung

**Mê cung sinh bằng DFS**
- DFS tạo ra mê cung ít nhánh, đường đi dài và "uốn lượn".
- **BFS, UCS, A\*** thăm nhiều nút vì phải kiểm tra nhiều đường đi tiềm năng.
- **DFS** thăm ít nút hơn vì đi theo một nhánh đến cùng, frontier nhỏ.
- **IDS** có chi phí cực cao do phải lặp lại nhiều mức độ sâu.

**Mê cung sinh bằng Prim**
- Prim tạo ra mê cung nhiều nhánh hơn, dẫn đến max_frontier của BFS và UCS lớn.
- DFS khi thăm theo nhánh vẫn giữ frontier nhỏ hơn nhiều, giải thích cho giá trị max_frontier thấp hơn so với BFS.
- **A\*** và **GBFS** thăm ít nút hơn BFS, nhưng max_frontier có thể tăng nếu heuristic dẫn sai hướng nhánh đông.

**Mê cung sinh bằng Kruskal**
- Mê cung Kruskal thường có nhiều đường rẽ nhánh nhưng ít "đường dài liên tục" như DFS.
- **BFS và UCS** thăm nhiều nút nhưng tổng thể ổn định hơn so với Prim, max_frontier vừa phải.
- **DFS** có số nút thăm lớn hơn DFS sinh ra bởi DFS do phải quay lui nhiều hơn trong mê cung rẽ nhánh.

##### Kết luận
- **DFS**: Frontier nhỏ, ít nút thăm trên mê cung ít nhánh, không đảm bảo đường ngắn nhất.
- **BFS/UCS**: Đảm bảo đường đi ngắn nhất, frontier lớn trên mê cung nhiều nhánh, tốn nút thăm.
- **IDS**: Tốn kém nhất về total_nodes, chi phí lặp lại nhiều lần.
- **DLS**: Nhanh, frontier nhỏ, nhưng không đảm bảo tìm được đường đi.
- **A\***: Cân bằng giữa BFS và DFS, tiết kiệm total_nodes nhờ heuristic, nhưng frontier có thể lớn khi heuristic dẫn sai nhánh.
- **GBFS**: Frontier trung bình, thăm ít nút hơn BFS nhưng không đảm bảo đường ngắn nhất.

### Đề xuất mở rộng
**IDA* (Iterative Deepening A*)**

**Mục tiêu:**  
Tìm đường đi tối ưu từ ô bắt đầu `(0,0)` đến ô đích `(rows-1, cols-1)` trong mê cung, đồng thời tiết kiệm bộ nhớ so với A*.

**Nguyên tắc hoạt động:**  
1. Sử dụng **DFS lặp theo ngưỡng f-limit**, với \( f = g + h \):
   - \( g \) = chi phí từ ô bắt đầu đến ô hiện tại (số bước đi).  
   - \( h \) = heuristic (Manhattan distance đến đích).
2. Khởi tạo **threshold = f(start)**.
3. Thực hiện DFS nhưng chỉ mở rộng các ô có \( f \leq \) threshold.
4. Nếu chưa tìm thấy đích, tăng threshold lên giá trị f nhỏ nhất vượt quá threshold trước đó và lặp lại.

**Heuristic sử dụng:**  
- Manhattan distance:
\[
h((r,c), goal) = |r - goal_r| + |c - goal_c|
\]

**Ưu điểm:**  
- Tiết kiệm bộ nhớ so với A*.  
- Đảm bảo tìm đường đi tối ưu nếu heuristic **admissible**.

**Nhược điểm:**  
- Thời gian có thể lâu hơn A* do lặp lại DFS nhiều lần.

### Cập nhật phân công, khối lượng công việc
<!-- công việc của các thành viên, tỷ lệ đóng góp của các thành viên -->
- Nguyễn Quốc Anh (20224919) (49 %): 
    - Cài đặt thuật toán sinh mê cung: Kruskal, Prim
    - Cài đặt thuật toán tìm đường trong mê cung: IDS, BFS, A*
    - Chỉnh sửa giao diện
    - Cài đặt thống kê và trực quan hóa kết quả
    - Tham gia viết và tổng hợp báo cáo
    - Tổng hợp code
- Cao Hữu Phúc (20235398) (17 %):
    - Cài đặt thuật toán sinh mê cung: DFS
    - Cài đặt thuật toán tìm đường trong mê cung: UCS
    - Tham gia viết báo cáo
- Nguyễn Trương Ngọc Mai (20235370) (17 %):
    - Cài đặt thuật toán tìm đường: DFS
    - Thiết kế và cài đặt giao diện
    - Tham gia viết báo cáo
- Đỗ Lê Hoàng (202416208) (17 %): 
    - Cài đặt thuật toán tìm đường trong mê cung: DLS, Greedy Best-First Search
    - Tham gia viết báo cáo
    - Đề xuất IDA*
    - Thiết kế và chuẩn bị slide thuyết trình
