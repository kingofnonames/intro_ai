## 3. Tiến độ giữa kỳ (W8)

### Chương trình
#### 1. Giới thiệu
Chương trình mô phỏng **tìm đường trong mê cung** với hai phần chính:

- **Sinh mê cung:** sử dụng thuật toán **DFS**.  
- **Tìm đường:** cài đặt các thuật toán **DFS, BFS, DLS, IDS, UCS**.

#### 2. Thuật toán sinh mê cung

##### DFS (Depth-First Search)
- Duyệt sâu từ ô bắt đầu, chọn ngẫu nhiên một ô kề chưa thăm, phá tường và đi tiếp.  
- Khi không còn ô chưa thăm → quay lui.  
- Kết quả: mê cung có **một đường duy nhất giữa hai điểm**.

#### 3. Thuật toán tìm đường đã cài đặt

##### DFS (Depth-First Search)
- Sử dụng **ngăn xếp (Stack)** để duyệt sâu.  
- Tìm đường nhưng **không đảm bảo ngắn nhất**.

##### BFS (Breadth-First Search)
- Sử dụng **hàng đợi (Queue)** để duyệt theo tầng.  
- Đảm bảo **đường đi ngắn nhất** đến đích.

##### DLS (Depth-Limited Search)
- Biến thể của DFS có **giới hạn độ sâu**.  
- Tránh lặp vô hạn, nhưng có thể **bỏ lỡ nghiệm nếu ở sâu hơn giới hạn**.

##### IDS (Iterative Deepening Search)
- Gọi **DLS nhiều lần**, mỗi lần tăng giới hạn độ sâu.  
- Kết hợp ưu điểm của **DFS (bộ nhớ ít)** và **BFS (đảm bảo tối ưu)**.

##### UCS (Uniform Cost Search)
- Duyệt theo **chi phí nhỏ nhất** (dùng hàng đợi ưu tiên).  
- Đảm bảo **đường có chi phí tối thiểu**, phù hợp khi ô có trọng số khác nhau.

#### 4. Cấu trúc chương trình
1. Sinh mê cung bằng **DFS**.  
2. Chọn thuật toán tìm đường (DFS, BFS, DLS, IDS, UCS).  
3. Hiển thị quá trình duyệt và kết quả trực quan.

**Nhận xét**

Do tất cả các cạnh trong mê cung đều có trọng số bằng 1, UCS hoạt động tương đương BFS, thể hiện qua các chỉ số gần như trùng nhau về số node duyệt, độ sâu tối đa và số node lưu trữ lớn nhất ở mọi kích thước mê cung. Với mê cung sinh bằng thuật toán DFS có ít nhánh rẽ, BFS và UCS duy trì frontier nhỏ và ổn định, cho thấy lợi thế về kiểm soát bộ nhớ so với DFS. Ngược lại, DFS thường đi sâu theo các hành lang dài, dẫn đến độ sâu tối đa và frontier lớn hơn khi kích thước mê cung tăng. IDS phát sinh chi phí dư thừa lớn do phải lặp lại tìm kiếm nhiều lần trong mê cung có độ sâu lớn, trong khi DLS bị giới hạn bởi độ sâu đặt trước và không phản ánh đúng độ khó của bài toán.

#### Vấn đề gặp phải
- **Chưa rõ cần thống kê những thông số nào** (ví dụ: số nút duyệt, thời gian chạy, độ sâu,...).    


#### 1. DFS (Depth-First Search)
**Ưu điểm:**
- Dễ cài đặt, sử dụng **ngăn xếp (stack)**.
- Tiết kiệm bộ nhớ vì chỉ lưu các nút trên đường đi hiện tại.
- Tốc độ nhanh khi đường đi nằm gần gốc hoặc sâu liên tục.

**Nhược điểm:**
- **Không đảm bảo tìm đường ngắn nhất.**
- Có thể **rơi vào vòng lặp vô hạn** nếu không đánh dấu đã thăm.
- Khi **map lớn**, dễ bị đi lạc sâu và tốn thời gian quay lui.

**Độ phức tạp:**
- Thời gian: `O(b^d)`  
- Bộ nhớ: `O(b*d)`  

### 2. BFS (Breadth-First Search)
**Ưu điểm:**
- Duyệt theo tầng → **luôn tìm được đường đi ngắn nhất**.
- Thích hợp khi chi phí các cạnh bằng nhau.

**Nhược điểm:**
- **Tốn bộ nhớ lớn** vì phải lưu toàn bộ nút trong một tầng.
- Khi **map lớn**, có thể chậm do số lượng nút tăng theo `b^d`.

**Độ phức tạp:**
- Thời gian: `O(b^d)`  
- Bộ nhớ: `O(b^d)`  

#### 3. DLS (Depth-Limited Search)
**Ưu điểm:**
- Giống DFS nhưng có **giới hạn độ sâu** → tránh vòng lặp vô hạn.
- Giảm chi phí thời gian so với DFS trong map rộng.

**Nhược điểm:**
- Có thể **bỏ lỡ nghiệm** nếu giới hạn độ sâu quá nhỏ.
- Cần chọn **độ sâu phù hợp** để đạt hiệu quả.

**Độ phức tạp:**
- Thời gian: `O(b^l)` (với `l` là giới hạn độ sâu)  
- Bộ nhớ: `O(b*l)`  

#### 4. IDS (Iterative Deepening Search)
**Ưu điểm:**
- Kết hợp ưu điểm của **BFS** (tối ưu đường đi) và **DFS** (tiết kiệm bộ nhớ).
- Phù hợp khi **không biết độ sâu nghiệm**.
- **Tốt khi map lớn** vì bộ nhớ chỉ tỷ lệ theo độ sâu.

**Nhược điểm:**
- **Tốn thời gian hơn BFS** do phải lặp lại tìm kiếm nhiều lần.

**Độ phức tạp:**
- Thời gian: `O(b^d)`  
- Bộ nhớ: `O(b*d)`  

#### 5. UCS (Uniform Cost Search)
**Ưu điểm:**
- Duyệt theo **chi phí nhỏ nhất** → tìm được **đường có tổng chi phí tối ưu**.
- Hoạt động tốt khi các ô có **trọng số khác nhau**.

**Nhược điểm:**
- **Chậm hơn BFS** khi trọng số bằng nhau.
- Cần **bộ nhớ lớn** để lưu trữ hàng đợi ưu tiên (priority queue).

**Độ phức tạp:**
- Thời gian: `O(b^{1 + ⌊C*/ε⌋})`  
- Bộ nhớ: tương tự thời gian.  

#### So sánh tổng quan

| Thuật toán | Đảm bảo tối ưu | Bộ nhớ | Thời gian | Hiệu quả khi map lớn |
|-------------|----------------|---------|------------|----------------------|
| **DFS** | Không | Thấp | Nhanh (khi sâu ít) | Kém (dễ đi lạc) |
| **BFS** | Có | Rất cao | Chậm | Không hiệu quả |
| **DLS** | Không | Thấp | Trung bình | Tốt nếu giới hạn hợp lý |
| **IDS** | Có | Thấp | Trung bình | Hiệu quả nhất |
| **UCS** | Có | Trung bình - cao | Chậm | Tốt với trọng số |

#### Kết luận
- **BFS**: chọn khi cần đường ngắn nhất, map nhỏ.  
- **DFS/DLS**: khi cần tiết kiệm bộ nhớ, không cần tối ưu.  
- **IDS**: cân bằng giữa tốc độ và bộ nhớ, phù hợp cho **map lớn**.  
- **UCS**: khi mê cung có trọng số hoặc chi phí khác nhau.
